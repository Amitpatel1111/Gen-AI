# Bag of word

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({"text":["people watch Movie",
                         "people watch cricket",
                         "people write comment",
                          "i write comment"],"output":[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
bow=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
bow.toarray()
```

```python
cv.transform(['I watch cricket']).toarray()
```

```python
x=bow.toarray()
y=df['output']
```

# N-Grams

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(ngram_range=(2,2))
```

```python
bow=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```
